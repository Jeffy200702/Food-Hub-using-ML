#Question 13:The company wants to provide a promotional offer in the advertisement of the restaurants. The condition to get the offer is that the restaurants must have a rating count of more than 50 and the average rating should be greater than 4. Find the restaurants fulfilling the criteria to get the promotional offer

Group by 'restaurant_name' and calculate the rating count and average rating

```python
restaurant_stats = data.groupby('restaurant_name')['rating'].agg(['count', 'mean']).reset_index()
```

Rename the columns for clarity

```python
restaurant_stats.columns = ['restaurant_name', 'rating_count', 'average_rating']
```

Filter restaurants that fulfill the criteria

```python
promotional_restaurants = restaurant_stats[(restaurant_stats['rating_count'] > 50) & (restaurant_stats['average_rating'] > 4)]
print("Restaurants fulfilling the criteria for the promotional offer:")
print(promotional_restaurants)
```

#Question 14:The company charges the restaurant 25% on the orders having cost greater than 20 dollars and 15% on the orders having cost greater than 5 dollars. Find the net revenue generated by the company across all orders

Initialize a revenue column

```python
data['revenue'] = 0.0
```

Calculate revenue based on the cost of orders

```python
data.loc[data['cost_of_the_order'] > 20, 'revenue'] = data['cost_of_the_order'] * 0.25
data.loc[(data['cost_of_the_order'] <= 20) & (data['cost_of_the_order'] > 5), 'revenue'] = data['cost_of_the_order'] * 0.15
```

Calculate total revenue

```python
total_revenue = data['revenue'].sum()
print(f"Total net revenue generated by the company across all orders: ${total_revenue:.2f}")
```

#Question 17:What are your conclusions from the analysis? What recommendations would you like to share to help improve the business? (You can use cuisine type and feedback ratings to drive your business recommendations)

Function without recursion and Average rating by cuisine type

```python
def analyze_cuisine_and_ratings(data):
    avg_ratings_by_cuisine = data.groupby('cuisine_type')['rating'].mean().sort_values(ascending=False)
    print("Average Ratings by Cuisine Type:")
    print(avg_ratings_by_cuisine)
    plt.figure(figsize=(10, 6))
    sns.barplot(x=avg_ratings_by_cuisine.index, y=avg_ratings_by_cuisine.values, palette='viridis')
    plt.xticks(rotation=45)
    plt.title('Average Ratings by Cuisine Type')
    plt.ylabel('Average Rating')
    plt.xlabel('Cuisine Type')
    plt.show()
analyze_cuisine_and_ratings(data)
```
